```python
from time import time
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
def fib_ricorsiva(n):
    ' Implementazione Ricorsiva'
    if n == 0 or n == 1:
        return 1
    return fib_ricorsiva(n-1) + fib_ricorsiva(n-2)
```

```python
def fib_dp(n):
    'Implementazione basata su Programmazione Dinamica'
    if n == 0 or n == 1:
        return 1
    aux = [1,1]
    for i in range(2,n+1):
        aux.append(aux[-1]+aux[-2])
    return aux[-1]
```

```python
''' Esperimento'''

n = 30

tempi_ricorsiva,tempi_dp = [], []


for i in tqdm(range(n)):
    t0 = time()
    fib_ricorsiva(i)
    t1 = time()
    tempi_ricorsiva.append(t1-t0)



for i in range(n):
    t0 = time()
    fib_dp(i)
    t1 = time()
    tempi_dp.append(t1-t0)

```

```python
x = [i for i in range(n)]
plt.plot(x,tempi_ricorsiva,label = 'Ricorsione',color = 'red')
plt.plot(x,tempi_dp, label = 'Programmazione Dinamica')

plt.xlabel('n')
plt.ylabel('tempo di esecuzione - log scale')
plt.yscale('log')
plt.title('Confronto Tempi Esecuzione Fibonacci')
plt.legend()
plt.show()
```

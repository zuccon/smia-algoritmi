```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ranges
n_range = np.linspace(1, 100, 400)  # n from 1 to 100
c_range = np.linspace(1, 100, 400)  # c from 1 to 100

# Create a meshgrid for n and c
N, C = np.meshgrid(n_range, c_range)

# Calculate the inequality (4n^2 + 100n + 500) < cn^2
Z = 4*N**2 + 100*N + 500 - C*N**2

# Plot
plt.figure(figsize=(10, 6))
# Use contourf to show the inequality region; levels=[np.min(Z), 0]
# to highlight where the inequality holds (i.e., Z < 0)
contour = plt.contourf(N, C, Z, levels=[np.min(Z), 0], cmap='Greens')
plt.colorbar(contour)

plt.title('Regions where 4*N**2 + 100*N + 500 - C*N**2 < 0')
plt.xlabel('n')
plt.ylabel('c')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ranges
n_range = np.linspace(1, 100, 400)  # n from 1 to 100
c_range = np.linspace(1, 100, 400)   # c from 0 to 10

# Create a meshgrid for n and c
N, C = np.meshgrid(n_range, c_range)

# Calculate the inequality (4n^2 + 100n + 500) >= cn^2
# We're looking for the region where this is true
Z = 4*N**2 + 100*N + 500 - C*N**2


# Plot
plt.figure(figsize=(10, 6))
# Use contourf to show the inequality region; levels=[0, np.max(Z)]
# to highlight where the inequality holds
contour = plt.contourf(N, C, Z, levels=[0, np.max(Z)], cmap='Greens')
plt.colorbar(contour)

plt.title('Regions where 4*N**2 + 100*N + 500 - C*N**2 >= 0')
plt.xlabel('n')
plt.ylabel('c')
plt.grid(True)
plt.show()
```
